### Task 1 - load the data and find the missing values and replace it with required mean 

```python
import pandas as pd
```

```python
data = pd.read_csv("DS1_C4_S4_Test_Scores_Data_Practice.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data["gender"].value_counts()
```

```python
data["gender"] = data["gender"].fillna("Male")
```

```python
a = data["test1"].mean()
a
```

```python
data["test1"] = data["test1"].fillna(a)
```

```python
b = data["test2"].median()
b
```

```python
data["test2"] = data["test2"].fillna(b)
```

```python
data.isnull().sum()
```

Interpretation - Replace the missing value of gender with the highest gender and then test1 missing values with the mean value of test 1 and then test2 missing values with the median value , i have filled all the missing values with the corresponding values
