### Task 1 - Display the result using a bar graph 

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("DS1_C4_S5_Employee_Data_Practice.xlsx")
```

```python
df
```

```python
bar = df.groupby("Department")["Annual Salary ($)"].sum()
```

```python
pd.DataFrame(bar)
```

```python
bar.reset_index()
```

```python
bar.plot(kind ="bar")
plt.title('Department-wise Annual Salary')
plt.xlabel('Department')
plt.ylabel('Annual Salary ($)')
plt.show()
```

Interpretation - Here i have displayed the sum of salary for each department

### Task 4 

```python
b = df.groupby(by=["Department"])["Annual Salary ($)"].sum()
```

```python
pd.DataFrame(b)
```

```python
b.plot(kind= "pie")
plt.pie(b,autopct='%.2f%%')
plt.title("Sum of annual salary in each department")

plt.show()
```

Interpretation - Here i have displayed the ratio of sum salaries of all employees in each department by using pie chart
