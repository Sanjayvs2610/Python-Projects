### Task 3 - Panda series variable containing the sum scores of physics and mathematics

```python
total_score
```

### Task 4  - the cut-off score to qualify the round 1 is 120 marks out of maximum marks of 200. Both mathematics and physics 

```python
cut_off = pd.Series(total_score>=120)
```

```python
filtered_op = total_score[cut_off] 
```

```python
print(filtered_op)
```

Interpretation - The above are the cut off marks of students who can qualify to round 1

### Task 5 - Create a dataframe containing the sum of math and physics scores of all students,where the sum of scores exceeds 120  

```python
round1_qualifiers = pd.DataFrame(filtered_op,columns = ["round1_qualifiers"])
```

```python
round1_qualifiers
```

### Task 6 - Create a dataframe for students who qualify for round 2 

```python
total_marks = total_score.values
total_marks
```

```python
df = pd.DataFrame()
```

```python
df["student_id"]= id
df["maths_marks"] = maths_marks
df["physics_marks"] = physics_marks
df["total_marks"] = total_marks
```

```python
df
```

```python
filtered_op = df[(df['maths_marks'] >= 50) & (df['physics_marks'] >= 50) & (df['total_marks'] >= 120)]
filtered_op
```

```python
round2_qualifiers = pd.DataFrame(filtered_op['student_id'])
round2_qualifiers
```

Interpretation - The above students are qualified to the round 2
